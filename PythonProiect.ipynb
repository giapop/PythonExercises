```python

magazinBucuresti = {"creioane", "pixuri", "caiete", "cartuse", "hartie"}
magazinConcurenta = {"pixuri", "markere", "penare", "hartie"}

produse_comune = magazinBucuresti.intersection(magazinConcurenta)
print("Produse comune:", produse_comune)

produse_unice_magazin1 = magazinBucuresti.difference(magazinConcurenta)
print("Produse unice în magazinul 1:", produse_unice_magazin1)

produse_unice_magazin2 = magazinConcurenta.difference(magazinBucuresti)
print("Produse unice în magazinul 2:", produse_unice_magazin2)

toate_produsele = magazinBucuresti.union(magazinConcurenta)
print("Toate produsele disponibile:", toate_produsele)
```

```python

dictionarMagazine = {"Bucuresti":5, "Brasov":1, "Craiova":1, "Deva":1, "Ploiesti":3, "Sighisoara":1, "Targoviste":2}
def afisareNrMagazine():
    for oras, nrMagazine in dictionarMagazine.items():
        if nrMagazine < 2 :
         print(f"{oras} : {nrMagazine}")
afisareNrMagazine()

def verifica_stoc_disponibil(stoc_produse, produs):
    if produs in stoc_produse:
        if stoc_produse[produs] > 0:
            return f"Produsul {produs} este disponibil în stoc."
        else:
            return f"Produsul {produs} nu este disponibil în stoc."
    else:
        return f"Produsul {produs} nu este disponibil în stoc."

stoc_produse_Brasov = {"creioane": 50, "pixuri": 0, "caiete": 100, "hartie": 200}

rezultat = verifica_stoc_disponibil(stoc_produse_Brasov, "pixuri")
print(rezultat)

rezultat = verifica_stoc_disponibil(stoc_produse_Brasov, "creioane")
print(rezultat)

def calcul_stoc_total(stoc_produse):
    stoc_total = 0
    for cantitate in stoc_produse.values():
        if cantitate > 0:
            stoc_total += cantitate

    return stoc_total

stoc_total = calcul_stoc_total(stoc_produse_Brasov)
print(f"Stocul total de produse în Brașov de papetărie este: {stoc_total}")
```
